```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open("/content/ironman.jpeg")
```

```python
def show(img,name):
  plt.imshow(img)
  plt.title(name)
  plt.show()
```

```python
show(img,"orginal ironman")
```

## Resize-transformation

moving from high resolution to low resolution

```python
import torchvision.transforms as transform
```

```python
reszie_transformation = transform.Compose([
    transform.Resize(size=(100,100))
])
```

```python
cropped_img = reszie_transformation(img)
```

```python
show(cropped_img,"cropped ironman")
```

## Flipping Transformation

```python
fliping_transformation = transform.Compose([
    transform.RandomHorizontalFlip(p=1)
])
```

```python
flipped_img = fliping_transformation(img)
show(flipped_img,"Flipped ironman")
```

```python
vertical_fliping_transformation = transform.Compose([
    transform.RandomVerticalFlip(p=1)
])
```

```python
flipped_img = vertical_fliping_transformation(img)
show(flipped_img,"Vertical Flipped ironman")
```

# Cropping Transformation

```python
cropping_transformation = transform.Compose([
    transform.RandomCrop(size=(200,200))
])
```

```python
cropped_img = cropping_transformation(img)
show(cropped_img,"Cropped ironman")
```

## Rotate Transformation

```python
rotate_transformation = transform.Compose([
    transform.RandomRotation(degrees=50)
])
```

```python
rotated_img = rotate_transformation(img)
show(rotated_img,"Roated by 30 ironman")
```
